```python
import tensorflow as tf
import cv2 as cv
import numpy as np
import os
from matplotlib import pyplot as plt
import pickle
import random

print(tf.version.VERSION)
video_dir = '/home/ibrahim/Environments/AI/TF/Adaptive_Gamma_Correction/Videos/'
images_dir = '/home/ibrahim/Environments/AI/TF/Adaptive_Gamma_Correction/Dataset/'
current_path = os.chdir(images_dir)
print(os.getcwd())
```

```python
gamma_min = 1/1.85
gamma_max = 1/0.6 ###Smaller the denominator, brighter the lower threshold
labels = []
image_dataset = []
img_size = 30
count = 0
for videos in os.listdir(video_dir):
    print(videos)
    vidcap = cv.VideoCapture(video_dir+videos)
    success,image = vidcap.read()
    count = 0;
    while (count <= 200):
        #Image converted to grayscale
        grayFrame = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
        grayFrame = grayFrame/255.0
        #Round to 2 decimal places
        gamma = round(random.uniform(gamma_min,gamma_max),2)
        #gamma correction applied image ** gamma
        corrected_image = np.power(grayFrame, gamma)
        corrected_image = cv.resize(corrected_image,(img_size,img_size))
        #corrected_gamma is saved upto two decimal places
        corrected_gamma = round(1/gamma,2)
#         image_name = str(count) + videos + "_" + "{:.2f}".format(corrected_gamma)+ ".jpg"
#         cv.imwrite(image_name, corrected_image)     # save frame as JPEG file    
        
        labels.append(corrected_gamma)
        image_dataset.append(corrected_image)
        
        success,image = vidcap.read()
        print(gamma, corrected_gamma, count)
        count += 1
        
print("Done")
```

```python
labels = np.array(labels)
image_dataset = np.array(image_dataset)

print(labels.shape)
print(image_dataset.shape)

plt.imshow(image_dataset[0], cmap = 'gray')
plt.show()
print(labels[0])

```
