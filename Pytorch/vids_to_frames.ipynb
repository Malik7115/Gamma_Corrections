```python
import cv2 as cv
import os
import numpy as np
import pandas as pd
import csv

vids_path   = "/home/ibrahim/Projects/Datasets/HPO_Recording/Videos/"
images_path = "/home/ibrahim/Projects/Datasets/HPO_Recording/Images/"

os.chdir(vids_path)
count = 0
```

```python
f = open(images_path + 'Annotations.csv', 'w', newline='')
writer = csv.writer(f)

cap = cv.VideoCapture(vids_path + '1.avi')
frames = 0
name = '1_image_'

while cap.isOpened():
    frames += 1
    ret, frame = cap.read()
    
    gamma     = np.random.uniform(0,2)
    
    gray      = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    procFrame = gray
    procFrame = gray/255.0

    procFrame = procFrame**2.0
    # procFrame = procFrame*255.0

    # if (frames%30 == 0):
    #     cv.imwrite(images_path + name + str(count) + ".jpg", gray)
    #     writer.writerow([name + str(count)])

    cv.imshow('frame', gray)
    cv.imshow('proc frame', procFrame)
    print(gamma)

    count += 1
    size = gray.shape
    if cv.waitKey(1) == ord('q'):
        break

f.close()

cap.release()
cv.destroyAllWindows()

# print(size)
```
